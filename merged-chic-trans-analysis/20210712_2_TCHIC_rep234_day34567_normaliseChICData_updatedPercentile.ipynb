## Normalise the data by dividing counts for all genes of one cell by the max count for that cell

#### k27me3

```python
TSS_k27_normalised_95 = TSS_k27.copy()
for cell in TSS_k27_normalised_95:
    TSS_k27_normalised_95[cell] = TSS_k27_normalised_95[cell]/np.nanpercentile(TSS_k27_normalised_95[cell],95)
```

```python
TSS_k27_normalised_98 = TSS_k27.copy()
for cell in TSS_k27_normalised_98:
    TSS_k27_normalised_98[cell] = TSS_k27_normalised_98[cell]/np.nanpercentile(TSS_k27_normalised_98[cell],98)
```
