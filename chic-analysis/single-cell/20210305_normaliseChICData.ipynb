# import packages

```python
import warnings
warnings.simplefilter(action='ignore', category=FutureWarning)

from pandas.io.parsers import read_csv
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import matplotlib as mpl
import sys, os
mpl.rcParams.update({'figure.max_open_warning': 0})
```

# load datasets

```python
pathToData = '/Users/m.blotenburg/Documents/Projects/TCHIC/data/rep2/20210215_OUD5651_K27me3K4me3_day5_20201023fixed/'
```

Load the chic data.

```python
TSS_table_k4 = pd.read_csv(pathToData + 'k4me3/chic/K4me3_TSS_count_table.csv', index_col=(0,1,2,3), 
                        header= 0, low_memory=False)
TSS_table_k4 = TSS_table_k4.iloc[1:]
TSS_table_k4.index.names = ['reference_name','start','end', 'bname']
TSS_table_k4.head(2)
```

```python
TSS_k4 = TSS_table_k4.droplevel(['start', 'end', 'reference_name'], axis=0)
TSS_k4 = TSS_k4.fillna(0)
```

```python
TSS_k4.T.head(2)
```

Do the same for k27 TSS

```python
TSS_table_k27 = pd.read_csv(pathToData + 'k27me3/chic/K27me3_TSS_count_table.csv', index_col=(0,1,2,3), 
                        header= 0, low_memory=False)
TSS_table_k27 = TSS_table_k27.iloc[1:]
TSS_table_k27.index.names = ['reference_name','start','end', 'bname']
TSS_table_k27.head(2)
```

```python
TSS_k27 = TSS_table_k27.droplevel(['start', 'end', 'reference_name'], axis=0)
TSS_k27 = TSS_k27.fillna(0)
TSS_k27.T.head(2)
```

## Normalise the data by dividing counts for all genes of one cell by the max count for that cell

```python
TSS_k27_normalised = TSS_k27
for cell in TSS_k27:
    TSS_k27_normalised[cell] = TSS_k27[cell]/TSS_k27[cell].max()
```

```python
TSS_k27_normalised.head(2)
```

```python
TSS_k4_normalised = TSS_k4
for cell in TSS_k4_normalised:
    TSS_k4_normalised[cell] = TSS_k4_normalised[cell]/TSS_k4_normalised[cell].max()
```

```python
TSS_k4_normalised.head(2)
```

## Check if everything worked as it should by making some more QC plots

```python
fig, ax = plt.subplots(4,1,figsize = (10,16))
plt.suptitle("K27me3, mean and max cuts per cell for raw and normalised data") 
sns.histplot(TSS_k27.mean(), color="lightsalmon", ax=ax[0], kde = True)
sns.histplot(TSS_k27_normalised.mean(), color="orangered", ax=ax[1], kde = True)
sns.histplot(TSS_k27.max(), color="palevioletred", ax=ax[2],kde=True,bins=50 )
sns.histplot(TSS_k27_normalised.max(), color="red", ax=ax[3],kde=True,bins=2)

ax[0].set_xlabel("Mean cuts per cell")
ax[1].set_xlabel("Mean cuts per cell, normalised")
ax[0].set_ylabel("Number of cells")
ax[1].set_ylabel("Number of cells")

ax[2].set_xlabel("Max cuts per cell")
ax[3].set_xlabel("Max cuts per cell, normalised")
ax[2].set_ylabel("Number of cells")
ax[3].set_ylabel("Number of cells")

fig.show()
```

```python
fig, ax = plt.subplots(4,1,figsize = (10,16))
plt.suptitle("K27me3, mean and max cuts per gene for raw and normalised data") 
sns.histplot(TSS_k27.mean(axis=1), color="lightsalmon", ax=ax[0], kde = True)
sns.histplot(TSS_k27_normalised.mean(axis=1), color="orangered", ax=ax[1], kde = True)
sns.histplot(TSS_k27.max(axis=1), color="palevioletred", ax=ax[2],kde=True,bins=25)
sns.histplot(TSS_k27_normalised.max(axis=1), color="red", ax=ax[3],kde=True,bins=25)

ax[0].set_xlabel("Mean cuts per gene")
ax[1].set_xlabel("Mean cuts per gene, normalised")
ax[0].set_ylabel("Number of genes")
ax[1].set_ylabel("Number of genes")
#ax[0].set_yscale("log")
#ax[1].set_yscale("log")

ax[2].set_xlabel("Max cuts per gene")
ax[3].set_xlabel("Max cuts per gene, normalised")
ax[2].set_ylabel("Number of genes")
ax[3].set_ylabel("Number of genes")

fig.show()
```

```python
fig, ax = plt.subplots(4,1,figsize = (10,16))
plt.suptitle("K4me3, mean and max cuts per cell for raw and normalised data") 
sns.histplot(TSS_k4.mean(), color="lightgreen", ax=ax[0], kde = True)
sns.histplot(TSS_k4_normalised.mean(), color="mediumseagreen", ax=ax[1], kde = True)
sns.histplot(TSS_k4.max(), color="limegreen", ax=ax[2],kde=True,bins=10 )
sns.histplot(TSS_k4_normalised.max(), color="darkgreen", ax=ax[3],kde=True,bins=2)

ax[0].set_xlabel("Mean cuts per cell")
ax[1].set_xlabel("Mean cuts per cell, normalised")
ax[0].set_ylabel("Number of cells")
ax[1].set_ylabel("Number of cells")

ax[2].set_xlabel("Max cuts per cell")
ax[3].set_xlabel("Max cuts per cell, normalised")
ax[2].set_ylabel("Number of cells")
ax[3].set_ylabel("Number of cells")

fig.show()
```
